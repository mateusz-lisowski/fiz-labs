### Cel ćwiczenia

Celem ćwiczenia jest badanie zjawiska zmiany długości prętów metalowych w funkcji temperatury oraz  
doświadczalne wyznaczenie współczynnika ich rozszerzalności liniowej.

### Opis badanego zjawiska

Współczynnik rozszerzalności liniowej (α) jest miarą zmiany długości ciała stałego w stosunku do zmiany temperatury.  
Określa, o ile zmienia się długość ciała stałego, na jednostkę długości pierwotnej, w wyniku wzrostu lub spadku temperatury.

### Wzory i zależności



Gdzie:
- α - współczynnik rozszerzalności liniowej
- x - długość przedmiotu po zmianie temperatury
- x0 - długość początkowa
- dT - przyrost temperatury

### Schemat układu pomiarowego



```python
# Zimportowanie potrzebnych modułów
import pandas as pd
import matplotlib.pyplot as plt

# Ścieżka do katalogu z danymi
DATA_PATH = 'data'


# Funcja zwaracjąca ramkę DataFrame w zależności od podanego pliku CSV
def get_data_frame(file_name: str) -> pd.DataFrame:

    # Załadowanie danych z pliku CSV
    csv_data = pd.read_csv(f'{DATA_PATH}/{file_name}', sep=';')
    
    # Stowrzenie ramki danych 
    csv_df = pd.DataFrame(csv_data)

    # Podzielenie kolumny d (zmiana długosci) przez 100, tak by zamnić ją na milimetry
    csv_df['d'] /= 100 

    # Zwrócenie gotowej ramki
    return csv_df


# Funkcja wyświetlająca wykres o zadanym tytule z zadanej ramki
def plot_from_df(title: str, df: pd.DataFrame) -> None:

    # Wybranie potrzebnej kolumny
    plt.plot(df['t'], df['d'])

    # Ustawienie etykiet osi
    plt.xlabel('temperatura [st. C]')
    plt.ylabel('zmiana długości [mm]')
    plt.title(title)
    
    # Wyświetl wykres
    plt.show()


```

### Wykres zmiany długości pręta z aluminium względem wzrostu temperatury

```python
# Pobranie danych z pliku data1.csv
df1 = get_data_frame('data1.csv')

# Narysowanie wykresu zmiany długości od temperatury
plot_from_df('Zmiana długosci pręta z aluminium względem wzrostu temperaruty', df1)
```

### Wykres zmiany długości pręta z żelaza względem wzrostu temperatury

```python
# Pobranie danych z pliku data1.csv
df2 = get_data_frame('data2.csv')

# Narysowanie wykresu zmiany długości od temperatury
plot_from_df('Zmiana długosci pręta z żelaza względem temperatury', df2)
```
