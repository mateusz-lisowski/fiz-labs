Wspczynnik rozszerzalnoci liniowej () jest miar zmiany dugoci ciaa staego w stosunku do zmiany temperatury.  
Okrela, o ile zmienia si dugo ciaa staego, na jednostk dugoci pierwotnej, w wyniku wzrostu lub spadku temperatury.

### Wzory i zalenoci



Gdzie:

### Schemat ukadu pomiarowego



```python
# Zimportowanie potrzebnych moduw
import pandas as pd
import matplotlib.pyplot as plt

# cieka do katalogu z danymi
DATA_PATH = 'data'


# Funcja zwaracjca ramk DataFrame w zalenoci od podanego pliku CSV
def get_data_frame(file_name: str) -> pd.DataFrame:

    # Zaadowanie danych z pliku CSV
    csv_data = pd.read_csv(f'{DATA_PATH}/{file_name}', sep=';')
    
    # Stowrzenie ramki danych 
    csv_df = pd.DataFrame(csv_data)

    # Podzielenie kolumny d (zmiana dugosci) przez 100, tak by zamni j na milimetry
    csv_df['d'] /= 100 

    # Zwrcenie gotowej ramki
    return csv_df


# Funkcja wywietlajca wykres o zadanym tytule z zadanej ramki
def plot_from_df(title: str, df: pd.DataFrame) -> None:

    # Wybranie potrzebnej kolumny
    plt.plot(df['t'], df['d'])

    # Ustawienie etykiet osi
    plt.xlabel('temperatura [st. C]')
    plt.ylabel('zmiana dugoci [mm]')
    plt.title(title)
    
    # Wywietl wykres
    plt.show()


```

### Wykres zmiany dugoci prta z aluminium wzgldem wzrostu temperatury

```python
# Pobranie danych z pliku data1.csv
df1 = get_data_frame('data1.csv')

# Narysowanie wykresu zmiany dugoci od temperatury
plot_from_df('Zmiana dugosci prta z aluminium wzgldem wzrostu temperaruty', df1)
```

### Wykres zmiany dugoci prta z elaza wzgldem wzrostu temperatury

```python
# Pobranie danych z pliku data1.csv
df2 = get_data_frame('data2.csv')

# Narysowanie wykresu zmiany dugoci od temperatury
plot_from_df('Zmiana dugosci prta z elaza wzgldem temperatury', df2)
```
